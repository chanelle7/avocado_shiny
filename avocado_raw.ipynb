```python
import numpy as np
import pandas as pd

pd.set_option('display.float_format', lambda x: '%.2f' % x)

avocado = pd.read_csv('avocado.csv')
avocado = avocado.iloc[:,1:]
avocado.Date = pd.to_datetime(avocado.Date)

avocado.sample(10)
```

```python
avocado.dtypes
```

```python
avocado['Total Volume'] = avocado['Total Volume'].round(0)
avocado['4046']=avocado['4046'].round(0)
avocado['4225']=avocado['4225'].round(0)
avocado['4770']=avocado['4770'].round(0)
avocado['Total Bags']=avocado['Total Bags'].round(0)
avocado['Small Bags']=avocado['Small Bags'].round(0)
avocado['Large Bags']=avocado['Large Bags'].round(0)
avocado['XLarge Bags']=avocado['XLarge Bags'].round(0)
avocado.sample(4)
```

```python
avocado.describe()
```

```python
avocado['test_volume'] = avocado['4046']+avocado['4225']+avocado['4770']
```

```python
avocado['missing_volume'] = avocado['Total Volume'] - avocado['test_volume']
```

```python
avocado[['Total Volume','test_volume','missing_volume','4046','4225','4770']]
```

there are other PLU codes for other varieties of avocados hence the missing data.

4046 are small avocados
4225 are medium sized avocados
4770 are large avocados

http://www.bobbywires.com/plu-1.php?S=P&L=A&V=AVOCADOS

the total avocados are for individual avocados just like plu tags are for single avocados sold in that week

```python
avocado['bag_test'] = avocado['Small Bags'] + avocado['Large Bags'] + avocado['XLarge Bags']
avocado['missing_bag'] = avocado['Total Bags'] - avocado['bag_test']
```

```python
avocado[['Total Bags', 'bag_test','missing_bag','Small Bags','Large Bags','XLarge Bags']]
```

no missing info in bags sold, all the bags add up to total bags

```python
avocado.rename(columns={'AveragePrice':'Avg_Price',
                       'Total Volume':'Total_Volume',
                       '4046':'Small_4046',
                       '4225':'Medium_4225',
                       '4770':'Large_4770',
                       'missing_volume':'Other_PLU',
                       'bag_test':'Total_Bags',
                       'Small Bags':'Small_Bags',
                       'Large Bags':'Medium_Bags',
                       'XLarge Bags':'Large_Bags',
                       'type':'Type',
                       'year':'Year',
                       'region':'Region'},
              inplace=True)
```
